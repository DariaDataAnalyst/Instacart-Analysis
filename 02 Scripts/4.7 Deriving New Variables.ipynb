### 2. Create "price_label" column

```python
# Create a subset of first million rows
df = df_ords_prods_merge[:1000000]
```

```python
df.head()
```

```python
df.shape
```

```python
# Define a function for price labels

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply price range function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
#Function to check what the most expensive product
df['prices'].max()
```

```python
# Use loc() method on df subset to apply price range function
```

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

### 3. Create "busiest_day" column

```python
# Find frequency of order_dow
df_ords_prods_merge['order_dow'].value_counts(dropna = False)
```

```python
# Use for-loop to determine whether orders are on "busiest day" (0 = Saturday), "least busy" (4 = Wednesday), or "regularly busy" (other days of the week).

result = []

for value in df_ords_prods_merge["order_dow"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
# Create new column in ords_prods_merge to view results in context.

df_ords_prods_merge['busiest_day'] = result
```

```python
df_ords_prods_merge.head()
```

```python
df_ords_prods_merge['busiest_day'].value_counts(dropna = False)
```

### 4. Task procedures

```python
# (2) Create column busiest_days" from "orders_day_of_week" to identify two busiest days (0, 1), slowest days (4, 3), or regularly busy (all other days).

result_new = []

for value in df_ords_prods_merge["order_dow"]:
  if value == 0 or value == 1:
    result_new.append("Busiest days")
  elif value == 4 or value == 3:
    result_new.append("Slowest days")
  else:
    result_new.append("Regularly busy")
```

```python
result_new
```

```python
# Check output
df_ords_prods_merge.head()
```

```python
# (3) Create new 'busiest_days' column in ords_prods_merge to view results in context
df_ords_prods_merge['busiest_days'] = result_new
```

```python
df_ords_prods_merge['busiest_days'].value_counts(dropna = False)
```

```python
# (4) Check value counts in "order_hour_of_day" column
df_ords_prods_merge['order_hour_of_day'].value_counts()
```

```python
# Create for-loop if statement labeling periods of time as 
# Most orders for 10, 11, 14, 15, 13, 12, 16, 9, 17
# Fewest orders for 6, 0, 1, 5, 2, 4, 3
# Average orders for the rest

result_new_hour = []

for value in df_ords_prods_merge["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9, 17]:
    result_new_hour.append("Most orders")
  elif value in [6, 0, 1, 5, 2, 4, 3]:
    result_new_hour.append("Fewest orders")
  else:
    result_new_hour.append("Average orders")
```

```python
result_new_hour
```

```python
# Create new column "busiest_period_of_day" in ords_prods_merge
df_ords_prods_merge['busiest_period_of_day'] = result_new_hour
```

```python
# (5) Print the frequency for "busiest_period_of_day" column
df_ords_prods_merge['busiest_period_of_day'].value_counts(dropna = False)
```
